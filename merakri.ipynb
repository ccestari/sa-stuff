```python
import polars as pl
import glob
```

```python
filepaths = glob.glob("meraki-webhooks-to-s3*")
```

```python
lazy_frame = pl.scan_ndjson(filepaths)
df = lazy_frame.collect()
display(df)
```

```python
df = df.unnest('payload')
```

```python
display(df)
```

```python
df = df.unnest('alertData')
display(df)
```

```python
df = df.explode('triggerData')
display(df)
```

```python
df = df.unnest('triggerData')
display(df)
```

```python
df = df.unnest('trigger')
display(df)
```
