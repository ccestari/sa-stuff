```python
import os
import polars as pl
```

```python
## Get File Extension
def get_file_extension(file_path: str) -> str:
    return os.path.splittext(file_path)
```

```python
# Identify JSON Format 
def identify_json_format(json_file: str) -> str:
    with open(json_file, 'r') as data:
        data_string = data.read()
    
    try:
        # Try parsing as a single JSON object/array
        json.loads(data_string)
        return "JSON"
    except json.JSONDecodeError:
        # If it fails, try parsing line by line (NDJSON)
        lines = data_string.strip().split('\n')
        is_ndjson = True
        for line in lines:
            try:
                json.loads(line)
            except json.JSONDecodeError:
                is_ndjson = False
                break
        if is_ndjson and len(lines) > 1: # Ensure it's not just a single JSON object on one line
            return "NDJSON"
        else:
            return "Neither"
```

```python
def csv_to_df(file_path: str, **kwargs) -> pl.DataFrame:
    try:
        if not file_path:
            return f"No Filepath provided"
        else:
            df = pl.read_csv(file_path, **kwargs)
            return df
    except Exception as e:
        raise exception(f'There was an error running {__name__}: {e}')
```

```python
def json_to_df(file_path: str, **kwargs) -> pl.DataFrame:
    try:
        if not file_path:
            return f"No Filepath provided"
        else:
            df = pl.read_json(file_path, **kwargs)
            return df
    except Exception as e:
        raise exception(f'There was an error running {__name__}: {e}')
```

```python
def ndjson_to_df(file_path: str, **kwargs) -> pl.DataFrame:
    try:
        if not file_path:
            return f'No Filepath provided'
        else:
            df = pl.read_ndjson(file_path **kwargs)
            return df
    except Exception as e:
        print(f'There was an error running {__name__}: {e}')
```

```python
def excel_to_df(file_path: str, **kwargs) ->pl.DataFrame:
    try:
        if not file_path:
            return f"No Filepath provided"
        else:
            df = pl.read_excel(file_path, **kwargs)
            return df
    except e:
        raise exception(f'There was an error running {__name__}: {e}')
```

```python
def dataframe_factory(file_path: str, **kwargs) -> pl.DataFrame:
    try:
        file_extension = get_file_extension(file_path)
    
        match file_extension:
            case 'csv':
                df = csv_to_df(file_path, **kwargs)
            case 'json':
                json_format = identify_json_format(file_path)
                match json_format:
                    case 'JSON':
                        df = json_to_df(file_path, **kwargs)
                    case 'NDJSON':
                        df = ndjson_to_df(file_path, **kwargs)
                    case 'Neither':
                        raise ValueError("JSON File is not a valid json document")
            case 'xls' | 'xlsx':
                df = excel_to_df(file_path, **kwargs)
            case _:
                print(f'The extension type {file_extension} is not compatible with Polars DataFrames')
    except Exception as e:
        raise exception(f'There was an error running {__name__}: {e}')
```
